Есть два набора точек - *x_hw* и *y_hw*. В рамках домашней работы нужно
* визуализировать набор точек
* найти коэффициенты регрессии $w_0, w_1$ по шагам, как в уроке
* посчитать предсказание в виде $\hat{y} = w_0 + w_1x$ и визуализировать его вместе с точками *x_hw* и *y_hw*

```python
x_hw = [50, 60, 70, 100]
y_hw = [10, 15, 40, 45]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_hw,y_hw)
plt.show()
```

```python
import numpy as np
X = np.array([[1,x] for x in x_hw])
X
```

```python
Y = np.array([[y] for y in y_hw])
Y
```

```python
X_T_X = (X.T).dot(X)
X_T_X
```

```python
from numpy.linalg import inv
X_T_X_inv = inv(X_T_X)
X_T_X_inv
```

```python
w = X_T_X_inv.dot(X.T).dot(Y)
w
```

```python
print(f'w_1 = {w[0][0]}, w_2 = {w[1][0]}')
```

```python
margin = 20
X_min = X[:,1].min() - margin
X_max = X[:,1].max() + margin

X_support = np.linspace(X_min,X_max,100)

Y_pred = w[0][0] + w[1][0]*X_support
```

```python
plt.scatter(X[:,1],Y[:,0])
plt.plot(X_support,Y_pred,'red')
plt.show()
```
