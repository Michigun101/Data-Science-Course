Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
import pickle
import numpy as np

# данные получены с помощью функции make_classification 
with open('clustering.pkl', 'rb') as f:
    data_clustering = pickle.load(f)

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.6, min_samples=8).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
```

```python
import matplotlib.pyplot as plt
labels = db.labels_

plt.scatter(X[:, 0], X[:, 1], s=40, c=labels, marker='o', alpha=0.8, label='data')
plt.show()
```

```python
# Цвета по меткам кластера (label -1 означает "шум" / выброс)
labels = db.labels_

# Количество кластеров (шум не считаем)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f"Количество кластеров: {n_clusters}")

# Построение графика
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', s=40)
plt.title('DBSCAN clustering (min_samples=8)')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.show()
```
