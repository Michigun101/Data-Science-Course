# Чтение и запись данных

## Чтение данных из web-сервиса

```python
import urllib.request as req
import json
import pandas as pd
```

### Поиск идентификатора города

```python
url = ("https://api.open-meteo.com/v1/forecast?"
       "latitude=55.7558&longitude=37.6176&current_weather=true")

session = req.urlopen(url)
response = session.read().decode()
session.close()

response
```

```python
url = ("https://api.open-meteo.com/v1/forecast?"
       "latitude=55.7558&longitude=37.6176&current_weather=true")

with req.urlopen(url) as session:
    response1 = session.read().decode()
response1
```

```python

with req.urlopen(url) as session:
    response1 = session.read().decode()
    data = json.loads(response1)
data
```

```python
city = data['latitude']
city
```

### Получение истории погоды

```python
url = (
    "https://archive-api.open-meteo.com/v1/archive?"
    "latitude=55.7558&longitude=37.6176&"
    "start_date=2020-03-01&end_date=2020-03-01&"
    "hourly=temperature_2m,relative_humidity_2m"
)

with req.urlopen(url) as session:
    response = session.read().decode()
    data = json.loads(response)
data
```

```python
df = pd.read_json(response)
df.head()
```
